```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE
```

```python
X, y = make_classification(n_classes=2, class_sep=2, weights=[0.7,
0.3],n_informative=3, n_redundant=1, flip_y=0,n_features=5,
n_clusters_per_class=1, n_samples=1000, random_state=42)
```

```python
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
before_counts = Counter(y)
after_counts = Counter(y_resampled)
```

```python
x_labels = ["MEN (Before SMOTE)", "WOMEN (Before SMOTE)", "MEN (After SMOTE)", "WOMEN (After SMOTE)"]
x_positions = np.arange(len(x_labels)) # Generate equally spaced positions
```

```python
plt.bar(x_positions[:2], before_counts.values(), color=['blue','fuchsia'], width=0.4, label="Before SMOTE")
plt.bar(x_positions[2:], after_counts.values(), color=['blue','fuchsia'], width=0.4, label="After SMOTE")
plt.xticks(x_positions, x_labels, rotation=20) # Rotate labels for better readability
plt.ylabel("Number of samples")
plt.title("Class Distribution Before and After SMOTE")
plt.show()
```
